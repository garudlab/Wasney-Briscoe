```python
import os
import sys
import lz4.frame
import numpy as np
import pandas as pd
sys.path.append("/u/project/ngarud/michaelw/Diversity-Along-Gut/ConventionalMouse/scripts/postprocessing/postprocessing_scripts/")
import config
```

```python
snps_summary_path = "%ssnps/snps_summary.tsv" % (config.data_directory)
snps_summary_df = pd.read_csv(snps_summary_path, sep = "\t")
```

```python
full_species_list = sorted(snps_summary_df.species_id.unique().tolist())
good_species = []
for species in full_species_list:
    sample_count = snps_summary_df[snps_summary_df['species_id'] == species].shape[0]
    if sample_count <= 1:
        print(species)
    if sample_count >= 2:
        good_species.append(species)
```

```python
out_path = "%sspecies_snps.txt" % (config.metadata_directory)
with open(out_path, "w") as f:
    for species in good_species:
        f.write(str(species) + "\n")
```
